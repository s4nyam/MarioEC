```python
import matplotlib.pyplot as plt
import numpy as np

# Define the number of generations and experiments
generations = np.arange(1, 1001)
num_experiments = 5

# Simulated fitness data for Algorithm 1
fitness_algorithm1 = np.array([
    [300 + 20 * np.sin(2 * np.pi * i / 1000) + np.random.normal(0, 10) for i in range(1000)],
    [310 + 15 * np.sin(2 * np.pi * i / 1000) + np.random.normal(0, 10) for i in range(1000)],
    [320 + 10 * np.sin(2 * np.pi * i / 1000) + np.random.normal(0, 10) for i in range(1000)],
    [330 + 5 * np.sin(2 * np.pi * i / 1000) + np.random.normal(0, 10) for i in range(1000)],
    [340 + 0.1 * i + np.random.normal(0, 10) for i in range(1000)]
])

# Simulated fitness data for Algorithm 2
fitness_algorithm2 = np.array([
    [290 + 20 * np.sin(2 * np.pi * i / 1000) + np.random.normal(0, 10) for i in range(1000)],
    [295 + 15 * np.sin(2 * np.pi * i / 1000) + np.random.normal(0, 10) for i in range(1000)],
    [300 + 10 * np.sin(2 * np.pi * i / 1000) + np.random.normal(0, 10) for i in range(1000)],
    [305 + 5 * np.sin(2 * np.pi * i / 1000) + np.random.normal(0, 10) for i in range(1000)],
    [310 + 0.1 * i + np.random.normal(0, 10) for i in range(1000)]
])

# Calculate statistics for Algorithm 1
mean_fitness_algorithm1 = np.mean(fitness_algorithm1, axis=0)
min_fitness_algorithm1 = np.min(fitness_algorithm1, axis=0)
max_fitness_algorithm1 = np.max(fitness_algorithm1, axis=0)

# Calculate statistics for Algorithm 2
mean_fitness_algorithm2 = np.mean(fitness_algorithm2, axis=0)
min_fitness_algorithm2 = np.min(fitness_algorithm2, axis=0)
max_fitness_algorithm2 = np.max(fitness_algorithm2, axis=0)

# Create separate plots for Algorithm 1 and Algorithm 2
plt.figure(figsize=(12, 6))

# Plot for Algorithm 1
plt.subplot(1, 2, 1)
plt.plot(generations, mean_fitness_algorithm1, label='Algorithm 1', color='b')
plt.fill_between(generations, min_fitness_algorithm1, max_fitness_algorithm1, color='b', alpha=0.2)
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('Algorithm 1 Fitness Range')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(loc='best')

# Plot for Algorithm 2
plt.subplot(1, 2, 2)
plt.plot(generations, mean_fitness_algorithm2, label='Algorithm 2', color='r')
plt.fill_between(generations, min_fitness_algorithm2, max_fitness_algorithm2, color='r', alpha=0.2)
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('Algorithm 2 Fitness Range')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(loc='best')

# Adjust layout for the plots
plt.tight_layout()

# Display the plots
plt.show()
```
